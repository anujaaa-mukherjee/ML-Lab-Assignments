```python

#importing all the necessary libraries
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#reading the csv
df=pd.read_csv("Heart (1).csv")
```

```python
#printing the 1st 5 rows of the dataset
df.head()
```

```python
#to check if any column is having null values
df.isnull().sum()
```

```python
#filling the null values with the mean
df["Ca"].fillna(df["Ca"].mean(), inplace=True)
```

```python
#checking the unique values present in ChestPain column
df["ChestPain"].unique()
```

```python
#SCATTER PLOT

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

plt.figure(figsize=(8, 6))
sns.scatterplot(x="Age",y="MaxHR",hue="ChestPain",data=df,palette=["blue","green","yellow","red"])
plt.title("Scatter Plot")
plt.xlabel("Age")
plt.ylabel("MaxHR")
plt.show()
```

```python

df["ChestPain"].value_counts()
```

```python
#BAR CHART
import matplotlib.pyplot as plt  

df["ChestPain"].value_counts().plot(kind="bar", color="yellow")  
plt.title("Bar plot for ChestPain Frequency")  
plt.xlabel("ChestPain")  
plt.ylabel("Frequency")  
plt.show()
```

```python
lables = df["Thal"].unique()
print(lables)
```

```python
lables = ['fixed', 'normal', 'reversable']
```

```python
# PIE CHART
plt.pie(x=df["Thal"].value_counts(),labels=lables, autopct="%1.1f%%")
plt.title("Thallium Heart Scan Results")
plt.show()
```

```python
#implementing box plot

plt.figure(figsize=(8,6))
sns.boxplot(y="Age",data=df,color="green")
plt.title("Box plot showing the range of people present in the dataset")
plt.show()
```

```python
#Last 5 entries of the dataset
df.tail()
```

```python

#implementing Line chart

plt.figure(figsize=(24,8))
sns.lineplot(x="Age",y="RestBP",data=df,marker="o",linestyle="-",color="red")
plt.title("Line chart: Age vs RestBP")
plt.xlabel("Age")
plt.ylabel("RestBP")
plt.grid(True)
plt.show()
```
